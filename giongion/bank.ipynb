```python
import pandas as pd

df = pd.read_excel('sheet commercial bank Social reporting.xlsx', sheet_name='Commercial Banks 1 (2)')
df["Nr. Pages"] = pd.to_numeric(df["Nr. Pages"], errors="coerce")
df["score for details"] = pd.to_numeric(df["score for details"], errors="coerce")
df["Balance sheet total (2024)"] = pd.to_numeric(df["Balance sheet total (2024)"], errors="coerce")
df["Revenue"] = pd.to_numeric(df["Revenue"], errors="coerce")

# Optionally, convert 'Size' to a categorical type
df["Size"] = df["Size"].astype("category")

print(df.dtypes)
display(df.head())
```

```python

# Map Size to numeric values
size_map = {'Small': 1, 'Medium': 2, 'Large': 3}
df["Size_numeric"] = df["Size"].map(size_map)

# List of columns to check correlation (use Size_numeric)
cols = ["Nr. Pages", "score for details", "Balance sheet total (2024)", 
        "Revenue", "Size_numeric", "ESG", "ESG risk score", "Social.1"]

from itertools import combinations

correlations = []
for col1, col2 in combinations(cols, 2):
    corr = df[col1].corr(df[col2])
    correlations.append((col1, col2, corr))

# Display as a DataFrame
correlation_df = pd.DataFrame(correlations, columns=["Column 1", "Column 2", "Correlation"])
correlation_df = correlation_df.sort_values(by="Correlation", ascending=False)
display(correlation_df)

```

```python
correlation_df.to_excel("correlation_output.xlsx", index=False)
```
